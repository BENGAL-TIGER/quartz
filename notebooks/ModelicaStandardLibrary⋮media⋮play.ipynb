```modelica
loadModel(Modelica) 
```

```modelica
model ComponentWithPureFluid replaceable package Medium =
    Modelica.Media.Interfaces.PartialPureSubstance;
  Medium.BaseProperties medium;
  Real rho;
equation
  medium.p = 1.0e5;
  medium.T = 293.15;
  medium.d = rho;
  // medium.d, medium.h, medium.u, medium.MM are available
end ComponentWithPureFluid;
```

```modelica
model ComponentWithH2O
    extends ComponentWithPureFluid(redeclare package Medium = Modelica.Media.Water.StandardWater); 
end ComponentWithH2O;
```

```modelica
simulate(ComponentWithH2O, numberOfIntervals = 1)
```

```modelica
plot(rho)
```

```modelica
model test 
  parameter Real pressure=2e5, temperature=298.15; 
  package CO2 = Modelica.Media.IdealGases.SingleGases.CO2; 
  CO2.BaseProperties medium; 
  Real density; 
equation 
  medium.p = pressure; 
  medium.T = temperature; 
  density = medium.d; 
end test; 
```

```modelica
simulate(test)
```

```modelica
plot(density)
```
